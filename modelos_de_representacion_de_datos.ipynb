Representación estática

Escenario: Representación de un catálogo de biblioteca.

Tarea: Crear un programa que utilice una estructura de datos estática (como una lista o un diccionario) para almacenar información sobre los libros del catálogo de una biblioteca. El programa debe permitir a los usuarios:

Añadir libros nuevos: Incluir título, autor, ISBN y género. Buscar libros: Por título, autor o ISBN. Mostrar todos los libros: Con formato.

```python
# ---------------- Catálogo de Biblioteca ----------------
catalogo = {}  # Diccionario para almacenar información de los libros

# Función para añadir libros
def add_book(titulo, autor, isbn, genero):
    catalogo[isbn] = {
        'titulo': titulo,
        'autor': autor,
        'genero': genero
    }
    print("Libro añadido al catálogo.")

# Función para buscar libros por título, autor o ISBN
def search_book(consulta):
    resultados = []
    for isbn, datos in catalogo.items():  # Recorre cada libro en el catálogo
        if (consulta.lower() in datos['titulo'].lower() or
            consulta.lower() in datos['autor'].lower() or
            consulta == isbn):
            resultados.append((isbn, datos))
    return resultados

# Función para mostrar todo el catálogo
def display_catalog():
    if not catalogo:
        print("\nEl catálogo está vacío.")
    else:
        print("\n--- Catálogo de Libros ---")
        for isbn, datos in catalogo.items():
            print(f"ISBN: {isbn}, "
                  f"Título: {datos['titulo']}, "
                  f"Autor: {datos['autor']}, "
                  f"Género: {datos['genero']}")

# Función para editar un libro existente
def edit_book(isbn):
    if isbn in catalogo:
        print("\nDatos actuales del libro:")
        print(f"Título: {catalogo[isbn]['titulo']}")
        print(f"Autor: {catalogo[isbn]['autor']}")
        print(f"Género: {catalogo[isbn]['genero']}")

        nuevo_titulo = input("Nuevo título (dejar vacío para no cambiar): ")
        nuevo_autor = input("Nuevo autor (dejar vacío para no cambiar): ")
        nuevo_genero = input("Nuevo género (dejar vacío para no cambiar): ")

        if nuevo_titulo:
            catalogo[isbn]['titulo'] = nuevo_titulo
        if nuevo_autor:
            catalogo[isbn]['autor'] = nuevo_autor
        if nuevo_genero:
            catalogo[isbn]['genero'] = nuevo_genero

        print("Datos del libro actualizados con éxito.")
    else:
        print("No se encontró un libro con ese ISBN.")

# ---------------- Menú interactivo ----------------
while True:
    print("\n=== MENÚ DE BIBLIOTECA ===")
    print("1. Añadir libro")
    print("2. Buscar libro")
    print("3. Mostrar catálogo")
    print("4. Salir")
    print("5. Editar libro")

    opcion = input("Elige una opción (1-5): ")

    if opcion == "1":
        # Añadir libro
        titulo = input("Título: ")
        autor = input("Autor: ")
        isbn = input("ISBN: ")
        genero = input("Género: ")
        add_book(titulo, autor, isbn, genero)

    elif opcion == "2":
        # Buscar libro
        consulta = input("Introduce título, autor o ISBN para buscar: ")
        resultados = search_book(consulta)
        if resultados:
            print("\nResultados encontrados:")
            for isbn, datos in resultados:
                print(f"ISBN: {isbn}, Título: {datos['titulo']}, Autor: {datos['autor']}, Género: {datos['genero']}")
        else:
            print("No se encontraron libros que coincidan con la búsqueda.")

    elif opcion == "3":
        # Mostrar catálogo
        display_catalog()

    elif opcion == "4":
        # Salir del programa
        print("Saliendo del programa...")
        break

    elif opcion == "5":
        # Editar libro
        isbn = input("Introduce el ISBN del libro a editar: ")
        edit_book(isbn)

    else:
        print("Opción no válida. Por favor, elige una opción del 1 al 5.")
```

Dinámica

Escenario de representación: Simulación de una cuenta bancaria.

Tarea: Crear un programa que utilice una estructura de datos dinámica (como una lista) para representar el historial de transacciones de una cuenta bancaria. El programa debe permitir a los usuarios:

Depositar dinero: Agregue un registro de transacción con el importe y escriba "depósito". Retirar dinero: Agregue un registro de transacción con el importe y escriba "retiro". Ver historial de transacciones: Muestre todas las transacciones en orden cronológico. Consultar saldo: Calcule el saldo actual con base en todas las transacciones. Ejemplo:

```python
# Simulación de una cuenta bancaria con historial dinámico
transactions = []  # Lista para almacenar todas las transacciones


# Función para depositar dinero
def deposit(amount):
    transactions.append({'type': 'deposit', 'amount': amount})
    print(f"Depósito de {amount} realizado con éxito.")


# Función para retirar dinero
def withdraw(amount):
    transactions.append({'type': 'withdraw', 'amount': amount})
    print(f"Retiro de {amount} realizado con éxito.")


# Función para ver historial de transacciones
def view_history():
    if not transactions:
        print("\nNo hay transacciones registradas.")
    else:
        print("\n--- Historial de transacciones ---")
        for transaction in transactions:
            tipo = "Depósito" if transaction['type'] == 'deposit' else "Retiro"
            print(f"{tipo}: {transaction['amount']}")


# Función para consultar saldo actual
def check_balance():
    balance = 0
    for transaction in transactions:
        if transaction['type'] == 'deposit':
            balance += transaction['amount']
        else:
            balance -= transaction['amount']
    return balance


# ---------------- Menú interactivo ----------------
while True:
    print("\n=== MENÚ DE CUENTA BANCARIA ===")
    print("1. Depositar dinero")
    print("2. Retirar dinero")
    print("3. Ver historial de transacciones")
    print("4. Consultar saldo")
    print("5. Salir")

    opcion = input("Elige una opción (1-5): ")

    if opcion == "1":
        cantidad = float(input("Ingrese la cantidad a depositar: "))
        deposit(cantidad)

    elif opcion == "2":
        cantidad = float(input("Ingrese la cantidad a retirar: "))
        withdraw(cantidad)

    elif opcion == "3":
        view_history()

    elif opcion == "4":
        print(f"\nSaldo actual: {check_balance()}")

    elif opcion == "5":
        print("Saliendo del programa...")
        break

    else:
        print("Opción no válida. Intente de nuevo.")
```

Persistente

Escenario de representación: Creación de una aplicación de lista de tareas pendientes sencilla.

Tarea: Desarrollar un programa que almacene una lista de tareas pendientes de forma persistente mediante un archivo (por ejemplo, un archivo de texto o un archivo JSON). El programa debe permitir a los usuarios:

Añadir tareas: Añadir nuevas tareas a la lista. Marcar tareas como completadas: Actualizar el estado de las tareas. Ver la lista de tareas pendientes: Mostrar todas las tareas con su estado. Guardar y cargar la lista: Desde y hacia el archivo. Ejemplo (con un archivo de texto):

```python
# Aplicación de lista de tareas persistente
# Guarda y carga las tareas desde un archivo de texto

def load_todo_list(filename="todo.txt"):
    tasks = []
    try:
        with open(filename, "r") as file:
            for line in file:
                tasks.append(line.strip())
    except FileNotFoundError:
        pass  # Si el archivo no existe, se empieza con una lista vacía
    return tasks


def save_todo_list(tasks, filename="todo.txt"):
    with open(filename, "w") as file:
        for task in tasks:
            file.write(task + "\n")


def add_task(tasks, task):
    tasks.append(task)
    print("Tarea añadida con éxito.")


def mark_complete(tasks, task_index):
    if 0 <= task_index < len(tasks):
        tasks[task_index] = tasks[task_index] + " [Completada]"
        print("Tarea marcada como completada.")
    else:
        print("Índice inválido.")


def view_todo_list(tasks):
    if not tasks:
        print("\nNo hay tareas en la lista.")
    else:
        print("\n--- Lista de Tareas ---")
        for index, task in enumerate(tasks):
            print(f"{index + 1}. {task}")


# ---------------- Menú interactivo ----------------
todo_list = load_todo_list()

while True:
    print("\n=== MENÚ DE TAREAS ===")
    print("1. Añadir tarea")
    print("2. Marcar tarea como completada")
    print("3. Ver lista de tareas")
    print("4. Guardar y salir")

    opcion = input("Elige una opción (1-4): ")

    if opcion == "1":
        tarea = input("Escribe la nueva tarea: ")
        add_task(todo_list, tarea)

    elif opcion == "2":
        view_todo_list(todo_list)
        try:
            indice = int(input("Número de la tarea a marcar como completada: ")) - 1
            mark_complete(todo_list, indice)
        except ValueError:
            print("Debes ingresar un número válido.")

    elif opcion == "3":
        view_todo_list(todo_list)

    elif opcion == "4":
        save_todo_list(todo_list)
        print("Lista guardada. Saliendo del programa...")
        break

    else:
        print("Opción no válida. Intenta de nuevo.")
```

Simulado

Escenario de representación: Modelado de una intersección de tráfico.

Tarea: Crear un programa que simule una intersección con semáforos y coches. Usar objetos para representar coches y semáforos. El programa debe:

Crear coches: Con horarios de llegada e indicaciones aleatorios. Controlar semáforos: Cambiar las luces a intervalos regulares. Mover coches: Según las señales de los semáforos y la posición de los coches. Mostrar la intersección: Representar visualmente los coches y los semáforos (p. ej., mediante texto). Ejemplo (simplificado):

```python
import random
import time

# ==========================
# CLASES PRINCIPALES
# ==========================

class Coche:
    def __init__(self, id, direccion):
        self.id = id
        self.direccion = direccion
        self.posicion = 0  # 0 = esperando, >0 = avanzando

    def avanzar(self):
        self.posicion += 1

    def __str__(self):
        estado = "En espera" if self.posicion == 0 else f"Avanzando ({self.posicion})"
        return f"Coche {self.id} -> Dirección: {self.direccion}, Estado: {estado}"


class Semaforo:
    def __init__(self, direccion):
        self.direccion = direccion
        self.color = "rojo"

    def cambiar_color(self):
        self.color = "verde" if self.color == "rojo" else "rojo"

    def __str__(self):
        return f"Semáforo {self.direccion}: {self.color.upper()}"


# ==========================
# DATOS DEL SISTEMA
# ==========================

coches = []
semaforos = {
    "Norte": Semaforo("Norte"),
    "Sur": Semaforo("Sur"),
    "Este": Semaforo("Este"),
    "Oeste": Semaforo("Oeste")
}
contador_coches = 0


# ==========================
# FUNCIONES DE LA SIMULACIÓN
# ==========================

def crear_coche():
    global contador_coches
    contador_coches += 1
    direccion = random.choice(["Norte", "Sur", "Este", "Oeste"])
    coche = Coche(contador_coches, direccion)
    coches.append(coche)
    print(f"\n[+] Se creó {coche}")


def cambiar_semaforos():
    for semaforo in semaforos.values():
        semaforo.cambiar_color()
    print("\n[~] Los semáforos han cambiado de color.")


def mover_coches():
    if not coches:
        print("\n[!] No hay coches en la intersección.")
        return

    for coche in coches:
        semaforo = semaforos[coche.direccion]
        if semaforo.color == "verde":
            coche.avanzar()
            print(f"[>] {coche} avanzó.")
        else:
            print(f"[X] {coche} está detenido (Semáforo en ROJO).")


def mostrar_interseccion():
    print("\n===== ESTADO DE LA INTERSECCIÓN =====")
    for semaforo in semaforos.values():
        print(semaforo)
    for coche in coches:
        print(coche)
    print("=====================================")


# ==========================
# MENÚ INTERACTIVO
# ==========================

def menu():
    while True:
        print("\n=== MENÚ DE SIMULACIÓN ===")
        print("1. Crear coche aleatorio")
        print("2. Cambiar semáforos")
        print("3. Mover coches")
        print("4. Mostrar estado de la intersección")
        print("5. Simulación automática (tiempo real)")
        print("6. Salir")

        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            crear_coche()
        elif opcion == "2":
            cambiar_semaforos()
        elif opcion == "3":
            mover_coches()
        elif opcion == "4":
            mostrar_interseccion()
        elif opcion == "5":
            simulacion_automatica()
        elif opcion == "6":
            print("Saliendo de la simulación...")
            break
        else:
            print("Opción no válida, intente de nuevo.")


# ==========================
# SIMULACIÓN AUTOMÁTICA
# ==========================

def simulacion_automatica():
    print("\n[~] Iniciando simulación automática (5 ciclos)...\n")
    for i in range(5):
        print(f"\n--- CICLO {i+1} ---")
        if random.random() < 0.6:  # probabilidad de que aparezca un coche
            crear_coche()
        cambiar_semaforos()
        mover_coches()
        mostrar_interseccion()
        time.sleep(2)  # pausa para simular tiempo real


# ==========================
# EJECUCIÓN
# ==========================

menu()
```
