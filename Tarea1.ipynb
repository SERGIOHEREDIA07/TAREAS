```python
"""Programa de práctica para aplicar estilos de código en Python (PEP8)."""

def clasificar_edad(edad):
    """Clasifica la edad en categorías."""
    if edad < 18:
        return "menor de edad"
    elif 18 <= edad < 60:
        return "adulto"
    return "adulto mayor"


def main():
    """Función principal del programa."""
    nombre = input("Ingrese su nombre: ")

    try:
        edad = int(input("Ingrese su edad: "))
    except ValueError:
        print("Error: la edad debe ser un número entero.")
        return

    categoria = clasificar_edad(edad)
    print(f"\nHola {nombre}, usted es {categoria}.")


if __name__ == "__main__":
    main()
```

```python
"""Cálculo del área de un triángulo con estilo PEP8."""


def calcular_area_triangulo(base, altura):
    """Calcula el área de un triángulo a partir de su base y altura."""
    return (base * altura) / 2


def main():
    """Ejecuta el programa principal."""
    base = 5
    altura = 10
    area = calcular_area_triangulo(base, altura)
    print(f"El área del triángulo es: {area}")


if __name__ == "__main__":
    main()
```

```python
"""Filtrar números pares de una lista siguiendo PEP8."""


def obtener_pares(numeros):
    """Devuelve una lista con los números pares de la lista original."""
    pares = []
    for numero in numeros:
        if numero % 2 == 0:
            pares.append(numero)
    return pares


def main():
    """Función principal."""
    numeros = list(range(1, 11))
    pares = obtener_pares(numeros)
    print(f"Números pares: {pares}")


if __name__ == "__main__":
    main()
```

```python
"""Calculadora simple en Python con estilo PEP8."""


def calcular(num1, num2, operacion):
    """Realiza operaciones básicas (+, -, *, /)."""
    if operacion == "+":
        return num1 + num2
    if operacion == "-":
        return num1 - num2
    if operacion == "*":
        return num1 * num2
    if operacion == "/":
        return num1 / num2
    return "Operación no válida"


def main():
    """Ejecuta la calculadora."""
    resultado = calcular(102, 5, "+")
    print(f"Resultado: {resultado}")


if __name__ == "__main__":
    main()
```

```python
"""Contar las vocales en una cadena siguiendo PEP8."""


def contar_vocales(cadena):
    """Cuenta el número de vocales en la cadena."""
    contador = 0
    for caracter in cadena:
        if caracter.lower() in "aeiou":
            contador += 1
    return contador


def main():
    texto = "Holia mundoa"
    total_vocales = contar_vocales(texto)
    print(f"El número de vocales es: {total_vocales}")


if __name__ == "__main__":
    main()
```

```python
"""Verificar si un número es primo siguiendo PEP8."""


def es_primo(numero):
    """Devuelve True si el número es primo, de lo contrario False."""
    if numero < 2:
        return False
    for i in range(2, numero):
        if numero % i == 0:
            return False
    return True


def main():
    numero = 7
    if es_primo(numero):
        print(f"{numero} es número primo")
    else:
        print(f"{numero} no es número primo")


if __name__ == "__main__":
    main()
```

```python
def greet(name):
    """Greets the user with a personalized message.

    Args:
        name: The name of the user to greet.

    Returns:
        A string containing the greeting message.
    """

    if name:
        return f"Hello, {name}!"
    else:
        return "Hello, world!"


if __name__ == "__main__":
    user_name = input("Enter your name: ")
    greeting = greet(user_name)
    print(greeting)
```
